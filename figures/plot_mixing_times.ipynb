# Make a plot of mixing times $t_2$ as a function of the number of intermediates $K$ for various $\mathcal{P}_{\text{acc}}$ 

```python
import os, sys
sys.path.append('../scripts')
from mixing import *

from matplotlib import pyplot as plt

P_acc_values = [0.4, 0.1, 0.02]
s = '$\mathcal{P}_{\\text{acc}}$'
labels = [f'{s} = {P_acc}' for P_acc in P_acc_values]
print('labels', labels)

K_values = range(5,50)

plt.figure(figsize=(4,3))

for i in range(len(P_acc_values)):

    P_acc = P_acc_values[i]
    thislabel = labels[i]

    t2_values = []
    for K in K_values:
        m = MixingTime(P_acc, K)
        # print(P_acc, K, m.evals, m.t_k)
        t2_values.append( m.t_k[0] )
    plt.plot(K_values, t2_values, '-', label=thislabel)

plt.xlabel('number of intermediates $K$')
plt.ylabel('mixing time $t_2$')
plt.legend(loc='best')
# plt.yscale('log')

plt.tight_layout()
plt.savefig('mixing_times.pdf')
```
